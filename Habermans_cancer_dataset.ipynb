# Exploratory Data Analysis on Haberman's Datset

<h6>FEATURES:</h6> 
 1. age of patient (age),
 2. year of operation (year), 
 3. number of cancer cell found (node)

<h6>LABELS:</h6> 
 1. patient died within 5 year (status=1), 
 2. patient survived within next 5 year ( status=2 )

<h6><p>OBJECTIVE:</p></h6>
    Agar apne man se koi naya patient ka features bata dega(jo patient hai hi nahi dataset me). Based on the features, we have to analyze whether a patient was dead within 5 years or the patient     had survived.

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.close();
sns.set_style('whitegrid')
#Pair-plot for haberman dataset
sns.pairplot(haberman,hue='status',vars = ["age", "year", "nodes"],size=3)
plt.show()
```

<h4>Obervations:</h4>

       Hance majority of the points are overlapping each other and there is no proper separation
       between the points, so it is difficult to classify the status of patient using pair-plot.


```python
sns.set_style('whitegrid')
sns.FacetGrid(haberman,hue="status",size=5).map(sns.distplot,'nodes').add_legend()
plt.show()
```

<h4>Obeservations:</h4>

      1.Patient with nodes less than 4 have more survival rate
      2.Patient with nodes more than 10 have less survival rate

```python
#Persons having age greater than 77 died
```

```python
status1 = haberman.loc[haberman["status"] == 1];
status2 = haberman.loc[haberman["status"] == 2];
status2
```

```python
counts, bin_edges = np.histogram(status1['age'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)
```

```python
counts, bin_edges = np.histogram(status1['year'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)
```

```python
counts, bin_edges = np.histogram(status1['nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)
```

```python
# 100% of status1 patient have node less than30
#<h4>Observations:</h4> <p>82% of patient survived had node less than 3</p>
```

```python
counts, bin_edges = np.histogram(status2['nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)
```
